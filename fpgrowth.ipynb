```python
import pandas as pd

# carregamos o dataframe na variavel df
aces = pd.read_csv('acessibilidade_rs.csv')
aces.shape
```

```python
# eh pego so o que vai ser usado
aces = aces.iloc[:, 1:]
# aces = aces[['IN_ACESSIBILIDADE_INEXISTENTE', 'FEDERAL', 'ESTADUAL', 'MUNICIPAL', 'PRIVADA', 'URBANA', 'RURAL']]
```

<h3>FP-GROWTH

```python
# fazemos uma amostragem dos dados. a partir daqui, haverá o mesmo número de 
# escolas em localidades rurais e urbanas representadas

# primeiro agrupamos por rural. isso gera um tipo especial de dataframe
# com dois grupos, rural = True e rural = False
g = aces.groupby('RURAL')

# depois aplicamos uma funcao lambda
# essa funcao torna um dataframe x em uma amostra (sample) dele mesmo
# no qual o tamanho da amostra de cada parte é o tamanho do menor grupo
# a funcao reset_index torna ele um dataframe novamente e tira as outras colunas
g = g.apply(lambda x: x.sample(g.size().min()).reset_index(drop=True))

g.shape
```

```python
# parte do fp-growth
from mlxtend.frequent_patterns import fpgrowth, association_rules

# aqui é gerado um dataframe com o fpgrowth, salvo em fis
# esse dataframe contem apenas o suporte e alguns frequent itemsets
fis = fpgrowth(g, min_support=0.1, use_colnames=True)

display(fis)
```

```python
# agora sao geradas regras de associacao
# o threshold foi escolhido para que nao houvesse muitas regras
rules = association_rules(fis, metric="confidence", min_threshold=0.3)

# aqui as regras sao ordenadas por confianca e salvos no mesmo dataframe
rules.sort_values(by=['confidence'], inplace=True, ascending=False)
display(rules)
```
