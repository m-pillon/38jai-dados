```python
import pandas as pd

# carregamos o dataframe na variavel df
aces = pd.read_csv('acessibilidade_rs.csv')
```

```python
# eh pego so o que vai ser usado
aces = aces.iloc[:, 1:]
# aces = aces[['IN_ACESSIBILIDADE_INEXISTENTE', 'FEDERAL', 'ESTADUAL', 'MUNICIPAL', 'PRIVADA', 'URBANA', 'RURAL']]
```

<h3>FP-GROWTH

```python
# parte do fp-growth
from mlxtend.frequent_patterns import fpgrowth, association_rules

# aqui Ã© gerado um dataframe com o fpgrowth, salvo em fis
# esse dataframe contem apenas o suporte e alguns frequent itemsets
fis = fpgrowth(aces, min_support=0.05, use_colnames=True)

display(fis)
```

```python
# agora sao geradas regras de associacao
# o threshold foi escolhido para que nao houvesse muitas regras
rules = association_rules(fis, metric="confidence", min_threshold=0.25)

# aqui as regras sao ordenadas por confianca e salvos no mesmo dataframe
rules.sort_values(by=['confidence'], inplace=True, ascending=False)
display(rules)
```
